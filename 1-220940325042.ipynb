## Data preparation

```python
df_model = df.copy()
#Making a copy to create a model
```

```python
df_model.head()
```

```python
#The car_name column is useless for predictions, thus dropping it
df_model.drop(columns=['Car_Name'], axis=1, inplace=True)
```

```python
df_model.head()
```

```python
#Creating dummy variable for categorical data
dummies = pd.get_dummies(df.iloc[:, 5:-1])
```

```python
df_model = df_model.join(dummies)
df_model.drop(columns=['Fuel_Type', 'Seller_Type', 'Transmission'], axis=1, inplace=True)
```

```python
df_model
```

```python
#Seprating independent and dependent columns
X = df_model.drop(columns=['Selling_Price']).values
y = df_model['Selling_Price'].values
```

```python
#Normalizing the Data using Standard Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:, 0:4] = sc.fit_transform(X[:, 0:4])
```

```python
X[0]
```

```python
#Creating training and testing split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Model building and evaluation

```python
#Creating the model
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100, criterion='squared_error', n_jobs=1, random_state=0)
model = rfr.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
df_compare = pd.DataFrame(y_test, columns=['Actual'])
df_compare['predicted'] =  y_pred
```

```python
df_compare
```

## Result with error calculation

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
